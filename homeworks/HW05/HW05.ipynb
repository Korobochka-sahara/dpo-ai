```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.dummy import DummyClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, roc_auc_score, confusion_matrix, roc_curve, auc
import os
df = pd.read_csv("..\datasets\S05-hw-dataset.csv")

print("\nПервые 5 строк датасета:")
print(df.head())
print("\n Информация о датасете:")
print(df.info())
print("\nОписательные статистики:")
print(df.describe().round(2))
print("\nАнализ целевого признака 'default':")
target_percent = df['default'].value_counts(normalize=True)
print(target_percent.round(1))

```

Датасет содержит 3000 объектов и 17 признаков, включая целевой default. Пропущенных значений нет. Явных аномалий в данных не обнаружено. Распределение таргета показывает умеренный дисбаланс: 60% клиентов без дефолта и 40% с дефолтом.

```python
X = df.drop(columns=['default', 'client_id'])  # Удаляем таргет и client_id
y = df['default']  # Таргет

print("Размерность признаков X:", X.shape)
print("Размерность таргета y:", y.shape)

# 2. Простая проверка (опционально, но рекомендуется)
print("\nПроверка типов данных:")
print(X.dtypes)

# Проверка, что все признаки числовые
non_numeric = X.select_dtypes(exclude=['number']).columns
if len(non_numeric) == 0:
    print("Все признаки числовые")
else:
    print(f"Внимание: найдены нечисловые признаки: {list(non_numeric)}")

# Разделение данных
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.25,
    random_state=42,
    stratify=y  # Сохраняем пропорции классов
)

# Бейзлайн-модель (предсказывает самый частый класс)
baseline = DummyClassifier(strategy="most_frequent", random_state=42)
baseline.fit(X_train, y_train)

# Оценка
y_pred = baseline.predict(X_test)
y_pred_proba = baseline.predict_proba(X_test)[:, 1]  # Вероятности класса 1

accuracy_dumb = accuracy_score(y_test, y_pred)
roc_auc_dumb = roc_auc_score(y_test, y_pred_proba)

print("\nОценка DummyClassifier:")
print(f"Strategy: 'most_frequent' (всегда предсказывает самый частый класс)")
print(f"Accuracy на тесте: {accuracy_dumb:.3f}")
print(f"ROC-AUC на тесте: {roc_auc_dumb:.3f}")

# Что предсказывает модель
most_frequent_class = y_train.mode()[0]
most_frequent_proportion = y_train.value_counts(normalize=True).max()
print(f"\nМодель всегда предсказывает класс: {most_frequent_class}")
print(f"Доля этого класса в данных: {most_frequent_proportion:.1%}")
print(f"Accuracy равна этой доле: {most_frequent_proportion:.3f}")
```

Бейзлайн (DummyClassifier) всегда предсказывает самый частый класс (0 = "нет дефолта"), что даёт accuracy = 0.600 (доля этого класса в данных) и ROC-AUC = 0.500 (как случайное угадывание). Это минимальный порог: любая реальная модель должна превзойти эти значения, чтобы считаться полезной.

```python
# Создание пайплайна
pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("logreg", LogisticRegression(max_iter=1000, random_state=42))
])

# Перебор C (GridSearchCV)
param_grid = {'logreg__C': [0.01, 0.1, 1.0, 10.0]}
grid_search = GridSearchCV(pipe, param_grid, cv=5, scoring='roc_auc')
grid_search.fit(X_train, y_train)

print(f"Лучшие параметры: {grid_search.best_params_}")
print(f"Лучший ROC-AUC (кросс-валидация): {grid_search.best_score_:.3f}")

# Оценка лучшей модели на тесте
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)
y_pred_proba = best_model.predict_proba(X_test)[:, 1]

# Метрики
accuracy = accuracy_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred_proba)

print(f"\nОценка на тесте:")
print(f"Accuracy:  {accuracy:.3f}")
print(f"ROC-AUC:   {roc_auc:.3f}")
print(f"Confusion Matrix:\n{confusion_matrix(y_test, y_pred)}")

# ROC-кривая
fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='orange', lw=2, label=f'AUC = {roc_auc:.3f}')
plt.plot([0, 1], [0, 1], color='navy', linestyle='--', label='Случайная')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC-кривая')
plt.legend()
plt.grid(alpha=0.3)

# Сохранение графика
figures_dir = "figures"
os.makedirs(figures_dir, exist_ok=True)
figure_path = os.path.join(figures_dir, "roc_curve_logreg.png")
plt.savefig(figure_path, dpi=150)
plt.show()
```

```python
results = pd.DataFrame({
    'Модель': ['DummyClassifier', 'LogisticRegression'],
    'Accuracy': [accuracy_dumb, accuracy],
    'ROC-AUC': [roc_auc_dumb, roc_auc],
    'Лучший C': ['-', grid_search.best_params_['logreg__C']]
})

print("Сравнение моделей")
print(results.to_string(index=False))

# Выводим все значения C из GridSearch
cv_results = pd.DataFrame(grid_search.cv_results_)
c_values = cv_results[['param_logreg__C', 'mean_test_score']].sort_values('param_logreg__C')
print("\nВлияние параметра C на ROC-AUC (кросс-валидация):")
for _, row in c_values.iterrows():
    print(f"  C = {row['param_logreg__C']}: ROC-AUC = {row['mean_test_score']:.7f}")
```

Логистическая регрессия значительно превзошла бейзлайн по ROC-AUC (с 0.50 до ~0.86), но показала чуть меньший рост Accuracy (с ~0.58 до ~0.79). Это объясняется тем, что ROC-AUC оценивает способность разделять классы, а Accuracy просто считает правильные ответы.

Изменение регуляризации C влияло на качество: слишком сильная регуляризация (C=0.01) ухудшала результаты, а оптимальное значение C находилось в диапазоне 0.1-10.0.

Логистическая регрессия является неплохим выбором для этой задачи. Она значительно лучше разделяет классы, чем случайная модель, работает быстро и интерпретируемо и автоматически балансирует признаки.
